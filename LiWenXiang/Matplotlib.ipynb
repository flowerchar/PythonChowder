```python
# 一个最简单的图像
import matplotlib.pyplot as plt
import numpy as np
# %matplotlib inline
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
plt.plot(x, y)
'''
    numpy负责生成数据
    matplotlib负责画图
'''
```

## Matplotlib Architecture
 - ### Backend:主要处理把图显示到哪里和画到哪里（操作系统角度）
 - ### Artist:图像显示成什么样子(颜色尺寸风格)·
 - ### Scripting:pyplot, Python和API（用户角度）

```python
t = np.arange(0, 2 , 0.1)
s = np.sin(t*np.pi)
# plt.plot(t,s, 'r--',t*2, s, '--')
plt.plot(t,s,'g--',label='aaa')
plt.plot(t*2,s,'b--',label='bbb')
plt.plot(t, s, )
plt.title("this is a demo")
plt.xlabel("this is x")
plt.ylabel("this is y")
plt.legend()
```

## matplotlib之subplot

```python
x = np.linspace(0,5)
y1 = np.sin(np.pi*x)
y2 = np.sin(np.pi*x*2)
plt.plot(x, y1, 'b--', label='sin(pi*x)')
plt.ylabel('y1 value')
plt.plot(x, y2, 'r--', label='sin(pi*2x)')
plt.ylabel('y2 value')
plt.title('this is x-y value')
plt.xlabel("x value")
plt.legend()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1, 'b--')
plt.ylabel('y1')
plt.subplot(2,1,2)
plt.plot(x,y2,'r--')
plt.ylabel('y2')
plt.xlabel('x')
# 先行后列
plt.show()
```

```python
figure, ax = plt.subplots(2,2)
ax[0][0].plot(x, y1)
ax[0][1].plot(x, y2)
plt.show()
```

## Pandas绘图之Series

```python
import numpy as np
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt
```

```python
s1 = Series(np.random.randn(1000)).cumsum()
s2 = Series(np.random.randn(1000)).cumsum()
```

```python
np.random.randn(10)
```

```python
aa = Series([1,2,3,4,5])
```

```python
s1.plot(kind='line', grid=True, label='S1', title="It is series", style='--')
s2.plot(label='S2')
plt.legend()
plt.show()
# 因为1000太大，导致柱状图绘制效果不好，使用小数据更合适
```

## Pandas绘图之DataFrame

```python
df.plot(kind='bar')
```

```python
df.plot(kind='barh')
```
